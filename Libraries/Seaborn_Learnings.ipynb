```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

## Understanding Heatmaps

```python
data = {
    'temperature': [30 + random.randint(-10, 10) for i in range(20)],
    'humidity': [20 + random.randint(-20, 20) for i in range(20)],
    'wind_speed': [10 + random.randint(-10, 10) for i in range(20)]
}
df = pd.DataFrame(data)
```

```python
corr = df.corr()
print(corr)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
